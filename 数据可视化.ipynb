## 坐标轴设置

```python
# 时间刻度调整
import datetime
import matplotlib as mpl

start = datetime.datetime(2018,7,1)
stop = datetime.datetime(2018,8,1)
delta = datetime.timedelta(days = 1)

dates = mpl.dates.drange(start,stop,delta)
y = np.random.rand(len(dates))

fig2 = plt.figure()
ax1 = fig2.add_subplot(111)
ax1.plot_date(dates,y,linestyle = '-',marker = ' ')
```

```python
# 双轴
x = np.arange(1,11,0.1)
y1 = x*x
y2 = np.log(x)

# 创建图形
fig = plt.figure()
ax1 = fig.add_subplot(111)
plt.ylabel('y1')
ax2 = ax1.twinx() # 关键代码
plt.ylabel('y2',rotation = 270) # y 轴的标签不要搞错了

ax1.plot(x,y1)
ax2.plot(x,y2,'--r') # --r 表示 线形为虚线的红色线
```

###  条形图

-  bar 竖轴柱状图
- barh 横轴柱状图
- bottom 关键词
- ind-+width 并列的柱状图，需要计算好 x 轴或 y 轴并列的距离

```python
# 函数使用包
from matplotlib.ticker import FuncFormatter

x = np.arange(4)
money = [1.5e5,2.5e6,5.5e6,2.0e7]
def million(x,pos):
    return '$%.1fm' % (x * 1e-6)

fig,ax = plt.subplots()
formatters = FuncFormatter(million)
ax.yaxis.set_major_formatter(formatters)
ax.bar(x,money)
plt.xticks(x,('bill','peter','david','bom'))
```

```python
# 横向柱状图
x = np.arange(4)
money = [1.5e5,2.5e6,5.5e6,2.0e7]
def million(x,pos):
    return '$%.1fm' % (x * 1e-6)

fig,ax = plt.subplots()
formatters = FuncFormatter(million)
ax.xaxis.set_major_formatter(formatters)
plt.yticks(x,('bill','peter','david','bom'))
ax.barh(x,money)
```

```python
# 拼接柱状图
n = np.arange(5)
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
width = 5

# 创建柱状图
p1 = plt.bar(n,menMeans)
p2 = plt.bar(n,womenMeans,bottom = menMeans)
plt.xticks(n,('g1','g2','g3','g4','g5'))
plt.yticks(np.arange(0,81,10))
plt.legend((p1,p2),('men','women'))
```

```python
# 并列柱状图
n = np.arange(5)
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
width = 0.35

# 创建柱状图
p1 = plt.bar(n-width/2,menMeans,width)
p2 = plt.bar(n+width/2,womenMeans,width)
plt.xticks(n,('g1','g2','g3','g4','g5'))
plt.yticks(np.arange(0,81,10))
plt.legend((p1,p2),('men','women'))
```

### 散点直方图

```python
x = np.random.randn(200)
y = x + np.random.randn(200)*0.5

#定义坐标轴参数
margin_border = 0.1
width = 0.6
margin_between = 0.02
height = 0.2

#散点图参数
left_s = margin_border #距左边框距离
bottom_s = margin_border #距底部距离
height_s = width #散点图高度
width_s = width #散点图宽度

#顶部直方图参数
left_x = margin_border #距左边框距离
bottom_x = margin_border+width+margin_between #距底部距离 = 散点图距底部距离+散点图高度+散点图与直方图的间距
height_x = height #直方图高度
width_x = width #直方图宽度

left_y = margin_border+width+margin_between
bottom_y = margin_border
height_y = width
width_y = height

plt.figure(1,figsize=(8,8))
rect_s = [left_s,bottom_s,width_s,height_s]
rect_x = [left_x,bottom_x,width_x,height_x]
rect_y = [left_y,bottom_y,width_y,height_y]

# 开始按照定义的尺寸绘图
axScatter = plt.axes(rect_s)
axHisX = plt.axes(rect_x)
axHisY = plt.axes(rect_y)
axHisX.set_xticks([])#取消顶部直方图的x轴刻度显示
axHisY.set_yticks([])#取消右侧直方图的y轴刻度显示

#绘制散点图
axScatter.scatter(x,y)

#设定边界
bin_width = 0.25
xymax = np.max([np.max(np.fabs(x)),np.max(np.fabs(y))])#取出x与y中最大的绝对值
lim =int(xymax/bin_width+1) * bin_width
axScatter.set_xlim(-lim,lim)
axScatter.set_ylim(-lim,lim)

bins = np.arange(-lim,lim+bin_width,bin_width)

axHisX.hist(x,bins=bins)
axHisY.hist(y,bins=bins,orientation='horizontal')#水平作图

axHisX.set_xlim(axScatter.get_xlim())
axHisY.set_ylim(axScatter.get_ylim());
```
